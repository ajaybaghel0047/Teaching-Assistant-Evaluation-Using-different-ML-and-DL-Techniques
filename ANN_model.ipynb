```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

# Reading data
path = "/home/shruti/taedata.csv"
df = pd.read_csv(path, header=None)

# Adding feature names to the data
data = pd.DataFrame(df.values, columns=["englishSpeaker", "CourseInstructor", "Course", "Semester", "Classsize", "grade"])

# One-hot encoding of 'Course' and 'CourseInstructor'
data1 = data[["CourseInstructor", "Course"]]
data2 = pd.get_dummies(data1.Course, prefix='Course', drop_first=True)
data3 = pd.get_dummies(data1.CourseInstructor, prefix='CourseInstructor', drop_first=True)

# Creating binary features for 'englishSpeaker' and 'Semester'
data4 = pd.DataFrame(data['englishSpeaker'])
data4['englishSpeaker1'] = data4['englishSpeaker'].apply(lambda x: 0 if x == 1 else 1)

data5 = pd.DataFrame(data['Semester'])
data5['Semester1'] = data5['Semester'].apply(lambda x: 0 if x == 1 else 1)

# Concatenating the encoded features
data7 = data4[['englishSpeaker1']].join(data5[['Semester1']])
data8 = pd.concat([data2, data3], axis=1)

# Scaling 'Classsize' feature
data9 = pd.DataFrame(data['Classsize'])
sc = MinMaxScaler()
data9['classsize'] = sc.fit_transform(data9)

# Final feature set (X)
X = pd.concat([data9[['classsize']], data7, data8], axis=1)

# Target variable (y)
y = data['grade']
y = pd.get_dummies(y)  # One-hot encoding of the target variable

# Checking for missing or non-numeric values in the dataset
print(X.isnull().sum())  # Should be zero
print(y.isnull().sum())  # Should be zero

# Ensure X and y are converted to float32 type and are NumPy arrays
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train = np.array(X_train).astype(np.float32)
X_test = np.array(X_test).astype(np.float32)
y_train = np.array(y_train).astype(np.float32)
y_test = np.array(y_test).astype(np.float32)

# Initializing the ANN
classifier = Sequential()

# Adding the input layer and first hidden layer
classifier.add(Dense(units=26, kernel_initializer='he_uniform', activation='relu', input_dim=X_train.shape[1]))

# Adding the second hidden layer
classifier.add(Dense(units=26, kernel_initializer='he_uniform', activation='relu'))

# Adding the output layer (for 3 classes)
classifier.add(Dense(units=3, kernel_initializer='he_uniform', activation='softmax'))

# Compiling the ANN
opt = SGD(lr=0.2, momentum=0.7)
classifier.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

# Fitting the ANN to the Training set
model_history = classifier.fit(X_train, y_train, validation_split=0.1, batch_size=10, epochs=50)

# Evaluating the model
_, train_acc = classifier.evaluate(X_train, y_train, verbose=0)
_, test_acc = classifier.evaluate(X_test, y_test, verbose=0)
print('Train Accuracy: %.3f, Test Accuracy: %.3f' % (train_acc, test_acc))
```
